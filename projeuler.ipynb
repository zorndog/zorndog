```python
#Problem 4: A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers
# is 9009 = 91 x 99. Find the largest palindrome made from the product of two 3-digit numbers.

def is_palindrome(n):
    
    reverse = 0
    num = n
    while n != 0:
        digit = n % 10
        reverse = reverse*10 + digit
        n //= 10
    if((num - reverse) != 0):
        return False
    return True

def problem4():
    palindromes=[]
    for i in range(100,999):
        for j in range(100,999):
            if (is_palindrome(i*j)):
                palindromes.append(i*j)
    print(max(palindromes))
problem4()
print(max(x*y for x in range(100,1000) for y in range(100,1000) if str(x*y) == str(x*y)[::-1]))
# Also, using str[::1] we can write: print(max(x*y for x in range(100,1000) for y in range(100,1000) if x*y = str(x*y)[::1]))
```

```python
#Problem 12: What is the value of the first triangle number to have over five hundred divisors?
# n-th triangular number is n(n+1)/2. 

import numpy as np


def factors(n):
    number = 0
    for i in range(1,int(np.sqrt(n))+1):
        if n%i ==0:
            number +=2
        if i*i ==n:
            number -=1
    return number

T=1

for x in range(2,1000000):
    T+=x
    if(factors(T)>500):
        print(T)
        break
```

```python
#Problem 14: Which starting number, under one million, produces the longest collatz chain?
import numpy as np

#This solution is not mine. The following code uses memoization: a way to save the pair (input,output). 
# For example, if we've already calculated the output for 13 (which is 10) if during the calculation of a collatz sequence
# of a bigger number we get to 13, we can just add current count + count of 13. This makes it much more efficient rather
# than calculating all the sequences from the beginning.
has2={}

def collatz(x):
    count = 1
    temp = x  
    while temp > 1:
        if temp % 2 == 0:
            temp = int(temp/2)
            if temp in has2:  # calculate temp and check if in cache
                count += has2[temp]
                break                    
            else:
                count += 1
        else:
            temp = 3*temp + 1
            if temp in has2:            
                count += has2[temp]
                break
            else:
                count += 1

    has2[x] = count
    return count

num=0
greatest=0
for i in range(1000000):
    c=collatz(i)
    if num<c:
        num=c
        greatest=i
print(greatest)
```

```python
#Problem 20: Find the sum of the digits in the number 100!.


memo={}
def factorial(k):
    if k<2:
        return 1
    if k not in memo:
       memo[k] = k * factorial(k-1)
    return memo[k]

number = [int(d) for d in str(factorial(100))]
    
print(sum(number))

```

```python
#Problem 21: Evaluate the sum of all the amicable numbers under 10000
import numpy as np

def sum_divisors(n):
    total = 1
    for i in range(2, int(np.sqrt(n)+1)):
        if n%i==0:
            total +=i
            y = n // i
            if y>i:
             total +=y
    return total

def amicable_numbers(limit):
    for a in range(limit):
        b = sum_divisors(a)
        if a != b and sum_divisors(b) == a:
            yield a

print(sum(amicable_numbers(10000)))
```
